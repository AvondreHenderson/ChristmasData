```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import datetime as dt
import numpy as np
from scipy.stats import linregress

# Assign Data File Locations
#christmas_billboard_path = "Resources/Music/christmas_billboard_data.csv"
holiday_songs_path = "Resources/Music/Holiday_Songs_Spotify.csv"

# Read the data files into dataframes
#christmas_billboard_df = pd.read_csv(christmas_billboard_path, encoding="ISO-8859-1")
holiday_songs_df = pd.read_csv(holiday_songs_path, encoding="ISO-8859-1")
```

```python
holiday_songs_df.head()
```

```python
holiday_songs_df = holiday_songs_df.drop(['Unnamed: 0','danceability', 'energy', 'loudness','speechiness', 'acousticness', 'instrumentalness','liveness','valence','playlist_img','album_img'], axis = 1)
#holiday_songs_df = holiday_songs_df.columns.droplevel(-1)
holiday_songs_df
```

```python
artist_df = pd.DataFrame(holiday_songs_df.artist_name.unique())
artist_df.columns = ['artist_name']
artist_df
```

```python
artist_list = artist_df['artist_name'].tolist()
artist_list
```

```python
count = 0
artist_code = []
for index in range(len(artist_list)): 
    count += 1
    code = ""
    artist = artist_list[index]
    if count < 10:
        code = artist[0]+artist[1]+str("00")+str(count)
        
    elif count <100: 
        code = artist[0]+artist[1]+str("0")+str(count)       
    else:
        code = artist[0]+artist[1]+str(count)
        
    artist_code.append(code)
```

```python
for a in artist_code:
    print(a)
```

```python
artist_df['artist_code'] = artist_code
```

```python
artist_df
```

```python
holiday_songs_df = pd.merge(holiday_songs_df, artist_df, how="left", on="artist_name")
holiday_songs_df 
```
