```python
#Dependencies
import pandas as pd
```

```python
#Store filepaths into variables
file_one = "Resources/Sales_November_2019.csv"
file_two = "Resources/Sales_December_2019.csv"

# Read our Data files with the pandas library
NovemberDF = pd.read_csv(file_one)
DecemberDF = pd.read_csv(file_two)
```

```python
#November pandas dataframe
NovemberDF.head()
```

```python
#December pandas dataframe
DecemberDF.head()
```

```python
#Combine dataframes and concatenate
frames = [NovemberDF, DecemberDF]

result_df = pd.concat(frames)
result_df
```

```python
#Sorting the DataFrame by date
sorted_df = result_df.sort_values("Order Date")
sorted_df
```

```python
#Drop all rows with missing information
clean_df = sorted_df.dropna()
clean_df
```

```python
#Drop all rows containing unneeded information
NovDec_df = clean_df[clean_df["Order ID"].str.contains("Order ID")==False]
NovDec_df
```

```python
#Split address column into 3 distinct sections
SplitAddress= NovDec_df["Purchase Address"].str.split(",", expand=True)
SplitAddress[["City","State", "Zip"]] = SplitAddress[2].str.split(" ", expand=True)
SplitAddress = SplitAddress.drop(columns=[2])
SplitAddress  = SplitAddress.drop(columns=["City"])
SplitAddress2 = SplitAddress.rename(columns={1: 'City'})
SplitAddress3 = SplitAddress2.rename(columns={0: 'Address'})
SplitAddress3
```

```python
#ReInsert address with split columns
#Display final cleaned and transformed dataframe
UpdatedNovDec = pd.concat([NovDec_df, SplitAddress3], axis=1)
UpdatedNovDec = UpdatedNovDec.drop(columns=["Purchase Address"])
UpdatedNovDec
```
